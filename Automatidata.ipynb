```python
#Next is to see the number of trips where we have the different numbers of passengers 
taxi_trip_df["passenger_count"].value_counts()

```

```python
#Let us examine the highest income from each passenger types 
taxi_trip_df[["passenger_count", "tip_amount"]].groupby("passenger_count").sum()
```

```python
taxi_trip_df["month"] = taxi_trip_df["tpep_dropoff_datetime"].dt.month_name()
taxi_trip_df["day"] = taxi_trip_df["tpep_dropoff_datetime"].dt.day_name()
```

```python
#we can begin to look at the number of rides per month

monthly_trips = taxi_trip_df["month"].value_counts(ascending = True).reset_index()
monthly_trips
```

**month with the highest amount paid is March and the lest occured in July**

let us observe the monthly 

```python
plt.figure(figsize  = (6, 8));
plt.title("Monthly Trips")
sns.barplot(x = "month", y= "count", data = monthly_trips );
plt.xticks(ticks = monthly_trips["month"] , rotation = 45 );
```

Let us look at the daily rides and their total 

```python
daily_trip = taxi_trip_df["day"].value_counts(ascending = False).reset_index()
daily_trip
```

```python
plt.figure(figsize = (6, 8))
plt.title("daily count");
sns.barplot(x = "day", y = "count", data = daily_trip);
plt.xticks(daily_trip["day"], rotation = 45);
```

```python
taxi_trip_df.head()
```

```python
daily_amount_total = taxi_trip_df[["total_amount", "day"]].groupby("day").sum()
day_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday","Friday", "Saturday"]
daily_amount_total = daily_amount_total.reindex( index = day_order)
daily_amount_total
```

```python
plt.figure(figsize = (8, 6));
plt.title("Total MAount on Daily Basis")
sns.barplot(x = "day", y = "total_amount", data = daily_amount_total);
```

```python
monthly_index =  ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_total_revenue = taxi_trip_df[["total_amount", "month"]].groupby("month").sum().reindex(index = monthly_index)
monthly_total_revenue
```

```python
plt.figure(figsize = (12,8))
plt.title("Monthly Revenue")
sns.barplot(data = monthly_total_revenue, x = "month", y = "total_amount")
plt.xticks(monthly_index, rotation = 45);
```

Its time we begin to look at the total trip covered. We can use the trip distance and the drop off location

```python
#so that 

trip_do_location = taxi_trip_df[["DOLocationID", "trip_distance"]].groupby("DOLocationID").mean().sort_values(by = "trip_distance", ascending = True)
```

```python
trip_do_location
```

```python
#Let us see the number of pick locations and drop of locations 
drop_off_locations = taxi_trip_df["DOLocationID"].nunique()
pick_locations = taxi_trip_df["PULocationID"].nunique()
pick_locations
```

```python
drop_off_locations
```

```python
plt.figure(figsize = (14, 6));
plt.title("Trip Distance vs DO Locations")
ax = sns.barplot(x = trip_do_location.index, y = trip_do_location["trip_distance"], order = trip_do_location.index)
ax.set_xticklabels([]);
ax.set_xticks([]);
```
